```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://www.dcard.tw/'
api = 'service/api/v2/posts?popular=true'
url_popular = url + api
```

```python
posts = requests.get(url_popular).json()

for post in posts:
    try:
        id_ = post['id']
        print(id_)
        title = post['title']
        print(title)
        like = post['likeCount']
        print(like)
    except UnicodeEncodeError:
        print('UnicodeEncodeError')
```
